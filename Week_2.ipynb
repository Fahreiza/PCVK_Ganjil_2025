```python
# Load the ID card image
id_card_img = cv.imread("/content/unnamed.jpg")

# Make a copy of the image to draw on
id_card_img_covered = id_card_img.copy()

# Estimate the coordinates for "Religion" and "Marital Status" sections
# These coordinates are approximate and may need adjustment based on the actual image layout
religion_coords_pt1 = (300, 350) # Example coordinates (x1, y1)
religion_coords_pt2 = (700, 400) # Example coordinates (x2, y2)

marital_status_coords_pt1 = (300, 400) # Example coordinates (x1, y1)
marital_status_coords_pt2 = (700, 450) # Example coordinates (x2, y2)


# Draw a rectangle over the Religion section (using BGR color format)
cv.rectangle(id_card_img_covered, religion_coords_pt1, religion_coords_pt2, color=(0, 0, 255), thickness=-1) # Red color, filled rectangle

# Draw a rectangle over the Marital Status section (using BGR color format)
cv.rectangle(id_card_img_covered, marital_status_coords_pt1, marital_status_coords_pt2, color=(255, 0, 0), thickness=-1) # Blue color, filled rectangle


# Display the image with the covered sections
plt.imshow(cv.cvtColor(id_card_img_covered, cv.COLOR_BGR2RGB))
plt.title("ID Card with Religion and Marital Status Covered (Group 5)")
plt.axis('off') # Hide axes
plt.show()
```
